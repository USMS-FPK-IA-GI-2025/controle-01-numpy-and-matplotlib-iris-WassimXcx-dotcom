# Partie A — Importation du Dataset, Arrays, Premières Manipulations

### A1. Importation du dataset
Importez NumPy et Matplotlib.  
Chargez le fichier `iris.csv` avec `np.genfromtxt` (`delimiter=','`, `skip_header=1`).  
Stockez les données dans un array nommé `data` et affichez sa `shape`.


```python
# Votre code ici
import numpy as np
import matplotlib.pyplot as plt

# Importer iris dataset avec numpy

features = np.genfromtxt('Iris.csv',delimiter=',',skip_header=1,usecols=(0,1,2,3,4))
species = np.genfromtxt('Iris.csv',delimiter=',',skip_header=1,usecols=(5),dtype=str)

#Le méthode set_portions pour la désactivation de le notation exponentielle
np.set_printoptions(
    suppress=True,
    precision=3
)

print(features)
print(features.shape)
print(species)
print(species.shape)
```

### A2. Séparation des features et des espèces
Séparez :
- les 4 premières colonnes numériques dans un array `features`
- la dernière colonne (espèces) dans un array `species`
Utilisez le slicing.


```python
# Votre code ici
#L'étape de séparation (A2) est omise car features et species sont déjà isolés et correctement typés (numérique/string) lors de l'importation corrigée.
#Exclure la premiere colone ID
features = features[:,1:]
print(features)
print(species)
```

### A3. Indexing

```python
# Votre code ici
print(features[:5])
```

### A4. Extraction d’une colonne
Créez un array `sepal_length` contenant uniquement la première colonne de `features`.

```python
# Votre code ici
# Souvent la variable sepal_lenght contient la deuxiemme colonne de features avec index 1
sepal_length = features[:,0]
print(sepal_length)
print(sepal_length.size)
print(sepal_length.dtype)
```

# Partie B — Statistiques Basiques et Opérations Mathématiques

### B6. Moyenne par colonne

```python
# Votre code ici
moyenne = np.mean(features,axis=0)
print(moyenne)
```

### B7. Min et Max

```python
# Votre code ici

min,max = np.min(features[:,2]),np.max(features[:,2])
print(min,max)
```

### B9. Normalisation

```python
# Votre code ici
normalized_petals = features[:,2] / max
print(normalized_petals)
```

### B10. Somme cumulée

# Partie D — Visualisation avec Matplotlib (Plots Simples)

### D17. Line plot
Line plot des longueurs des sépales pour les 50 premières fleurs.


```python
# Votre code ici
plt.title('Longueur des sépales')
plt.plot(sepal_length[:50],)
plt.xlabel('Nombre des sépales')
plt.ylabel('Longueur')
plt.show()
plt.savefig('figures/sepal_length.png')
```

### D18. Scatter plot
Scatter plot : longueur vs largeur des pétales.


```python
# Votre code ici
plt.title('Longueur vs largeur des pétales')

species_colors = {
    'Iris-setosa': 'darkred',
    'Iris-versicolor': 'royalblue',
    'Iris-virginica': 'forestgreen'
}


colors_for_plot = [species_colors[s] for s in species]

plt.scatter(features[:,2],features[:,3],c=colors_for_plot)
plt.xlabel('Longueur')
plt.ylabel('Largeur')
plt.legend(
    ['Iris-setosa','Iris-versicolor','Iris-virginica']
    ,loc='upper left'
)
plt.show()
plt.savefig('figures/longvslargPetales')
```

### D19. Bar chart
